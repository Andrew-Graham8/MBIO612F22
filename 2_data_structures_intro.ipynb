#### Question 7
* Let's use a different strategy and impute the missing value.
  * replace the missing values for Solar.R using that month's average.
  * Example:
    * The missing value for line 6 should be replaced with the average for month 5.
    * The missing value for line 97 should be replaced with the average for month 8.

```r
df_5= airquality
May_data = subset(df_5, Month==5)
May_mean =mean(May_data$Solar.R, na.rm=TRUE)
May_data$Solar.R[is.na(May_data$Solar.R)] = May_mean
Aug_data = subset(df_5, Month==8)
Aug_mean =mean(Aug_data$Solar.R, na.rm=TRUE)
Aug_data$Solar.R[is.na(Aug_data$Solar.R)] = Aug_mean
June_data=subset(df_5, Month==6)
July_data=subset(df_5, Month==7)
Sept_data=subset(df_5, Month==9)
rbind(May_data,June_data,July_data,Aug_data,Sept_data)
```
